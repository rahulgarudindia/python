```python
#Find sum of even numbers from 1 to 10

x=1

total=0
while x<11:
    if(x%2==0):
    
        total=x+total
        print("total", total)
   
    x=x+1;
   
    
```

```python
##Find sum of squares of even numbers from 1 to 10

x=1
total=0
while(x<11):
    if(x%2==0):
       total=x*x+total
    print(x,"-------",total)
    
    x=x+1
```
