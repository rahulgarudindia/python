```python
### break
l=[1,20,25,4,8,20]
n=[]
for i in l:
  
    if i==4:
        break;
    n.append(i+5)

print(n)
```

```python
##continue
l=[1,20,25,4,8,20]
n=[]
for i in l:
   
    if i==4:
        print(i)
        print(n)
        continue;
        print(i)
        print(n)
    n.append(i+5)

print(n)
```

```python
### nested loops 
for i in [1,2]:
    for j in [4,5]:  
         print(i,j)
```
