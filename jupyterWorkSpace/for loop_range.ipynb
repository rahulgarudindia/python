```python
# for in sequence 

l=(1,2,3,4,5)

for i in l:
    print(l)


l=(1,2,3,[4,5])

for i in l:
    print(l)

    
```

```python
#### dictionary 

d={1:25,"a":90}

for i in d.items():
    print(i)
```

```python
## print only odd number 

l=[]
for i in range (0,101,2):
    l.append(i+1)
print(l)
```

```python
## print even number 
x=0
l=[]
while x<=100:
    if x%2==0:
        l.append(x)
    x=x+1
print(l)
```
